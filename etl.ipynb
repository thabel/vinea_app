```python
sales = pd.read_csv("data/sales_total_dropped.csv")
sales
```

```python
# Updating the supplier Id in the Items table

for i,row1 in suppliers.iterrows():
    items["id supplier"].where(items["id supplier"]!=row1["id"],i,inplace=True)
    orders["id supplier"].where(orders["id supplier"]!=row1["id"],i,inplace=True)

for i,row1 in items.iterrows():
    sales["id item"].where(sales["id item"]!=row1["id"],i,inplace=True)
    orders["id item"].where(orders["id item"]!=row1["id"],i,inplace=True)
```

```python
items["id supplier"] = items["id supplier"]+1
items
```

```python
orders["id supplier"] = orders["id supplier"]+1
orders["id item"] = orders["id item"]+1
orders
```

```python
sales["id item"] = sales["id item"]+1
sales
```

```python
suppliers.to_csv('data_corrected/suppliers.csv')
items.to_csv('data_corrected/items.csv')
orders.to_csv('data_corrected/orders.csv')
sales.to_csv('data_corrected/sales.csv')
```
